```python
# load libraries and packages

from google.cloud import bigquery
from sqlalchemy import create_engine
from IPython.display import display, HTML
import pandas as pd
```

```python
# set project id for BigQuery

project_id = 'woe-system'
```

```python
# set client for BigQuery

client = bigquery.Client(project=project_id)
```

```python
# Construct a BigQuery connection string
# Replace with your project ID, dataset ID, etc.
BQ_rdsds_connection_string = f'bigquery://woe-system/postgres_rds_public_datstream'
```

```python
# Create an engine using SQLAlchemy
engine = create_engine(BQ_rdsds_connection_string)
```

```python
# Your SQL query
sql_query = """
SELECT *
FROM woe-system.postgres_rds_public_datstream.public_biometrics_vital_summary_activity
LIMIT 10;
"""
```

```python
# Execute the query using pandas and SQLAlchemy
df = pd.read_sql_query(sql_query, engine)
```

```python
# Print or work with the DataFrame
display(df)
```
